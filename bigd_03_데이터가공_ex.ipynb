####12. dplyr패키지를 이용하여 DF에서 team(소속팀)을 group으로 묶어 team(소속팀)별 평균 G(출전게임수), 평균 AB(타수(타석-볼넷-사구-희생번트-희생플라이)), 합계 R(득점)을 출력하세요.

```r
#12번

DF %>%
  group_by(team) %>%
    summarise(
      mean_G = mean(G, na.rm = TRUE),
      mean_AB = mean(AB, na.rm = TRUE),
      sum_R = sum(R,na.rm = TRUE)) %>%
  show()
```

####13. dplyr패키지를 이용하여 DF에서 team(소속팀)과 batter_name(타자 이름)을 group으로 묶어 소속팀의 타자별 평균 G(출전게임수), 평균 AB(타수(타석-볼넷-사구-희생번트-희생플라이)), 합계 R(득점)을 출력하세요.

```r
#13번
DF %>%
  group_by(team,batter_name) %>%
    summarise(G_AVG = mean(G, na.rm = T),
              AB_AVG = mean(AB, na.rm = T),
              R_SUM = sum(R, na.rm = T))
```

####14. dplyr패키지를 이용하여 DF를 team(소속팀)은 오름차순으로, avg(타율)은 내림차순을 기준으로 정렬하여 team(소속팀), batter_name(타자 이름), avg(타율)만 출력하세요.

```r
#14번

DF %>%
  select(team, batter_name, avg) %>%
  arrange(team, desc(avg)) %>%
  show()
```

####16. dplyr패키지를 이용하여 DF에서 team(소속팀)이 ‘히어로즈’인 선수들의 team(팀), batter_name(타자 이름), avg(타율)을 avg(타율)이 높은 순으로 중복 없이 정렬하여 출력하세요.

####17. dplyr패키지를 이용하여 DF에서 batter_id(타자의 고유 아이디)는 ‘ID’, batter_name(타자 이름)은 ‘이름’, year(년도)는 ‘년도’라는 이름으로 변경하여 DF1에 저장한 후 DF1과 DF를 출력해 비교하세요.

#### 16. temp <- c(sample(1:10, 20, replace = T), -15:-11, 21:25, NA)로 temp를 만든 후 다음의 내용을 순서대로 작성하여 결과를 확인하세요.

A. temp에 저장된 데이터의 개수 출력

B. temp에 저장된 NA의 개수 출력

C. 이상값을 NA로 변경하여 temp_clear_outlier에 저장

D. temp_clear_outlier에 저장된 NA의 개수 출력

E. temp_clear_outlier에서 NA을 제거하여 temp_new에 저장

F. temp_new에 저장된 데이터의 개수 출력

```r
#Q no.16
temp <- c(sample(1:10, 20, replace = T), -15:-11, 21:25, NA)
temp
```

```r
#Q no.16A _ temp에 저장된 데이터의 개수 출력
length(temp)
```

```r
#Q no.16B _ temp에 저장된 NA의 개수 출력
cat("temp에 저장된 NA의 개수 : ", sum(is.na(temp)))
```

```r
#Q no.16C _ 이상값을 NA로 변경하여 temp_clear_outlier에 저장
boxplot(temp)
summary(temp)
```

```r
#IQR
temp_no_NA <- na.omit(temp)
iqr <- IQR(temp_no_NA)
#UF & LF 로 NA처리
temp_clear_outlier <- ifelse(temp_clear_outlier > summary(temp)[5] + iqr*1.5, NA, temp)
temp_clear_outlier <- ifelse(temp_clear_outlier < summary(temp)[2] - iqr*1.5, NA, temp)

temp_clear_outlier
```

```r
#Q no.16D _ temp_clear_outlier에 저장된 NA의 개수 출력
cat("temp_clear_outlier에 저장된 NA의 개수 : ", sum(is.na(temp_clear_outlier)))
```

```r
#Q no.16E _ temp_clear_outlier에서 NA을 제거하여 temp_new에 저장
temp_new <- na.omit(temp_clear_outlier)
temp_new
```

```r
#Q no.16E _ F temp_new에 저장된 데이터의 개수 출력
length(temp_new)
```

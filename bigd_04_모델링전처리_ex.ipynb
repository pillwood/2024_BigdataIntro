```r
#이 부분부터 중간고사 시험 범위

#8 TB 데이터에서 year(년도)가 2017년인 데이터들의 year, team, batter_name(타자이름), HR
# (홈런), OPS(OBP(출루율)+SLG(장타율))을 읽어 TB_2017에 저장한 후 전체 데이터 행 수와
# NA를 포함한 행 수를 확인합니다. 이후 TB_2017의 HR(홈런), OPS(OBP(출루율)+SLG(장타율))의
# boxplot()을 그려 outlier가 존재하는지 확인합니다.
TB_2017 <- TB %>%
  filter(year >= 2017) %>%
   select(year, team, batter_name, HR, OPS)

cat("데이터의 행수 : ", nrow(TB_2017),"\n")

cat("NA포함 행수 : ", table(is.na(TB_2017)),"\n")

TB_2017 %>% select(HR, OPS) %>% boxplot()
```

####9. [확인] HR_outlier에 TB_2017 데이터에서 team별 HR(홈런)이 outlier에 해당하는 선수들의 이름을 is_out 필드에 저장(outlier에 해당하지 않으면 NA)하는 방법을 확인하세요.

```r
#9 [확인] HR_outlier에 TB_2017 데이터에서 team별 HR(홈런)이 outlier에 해당하는 선수들의 이름을
#  is_out 필드에 저장(outlier에 해당하지 않으면 NA)하는 방법을 확인하세요.

# outlier 데이터를 확인(TRUE/FALSE)하는 함수
is_outlier <- function(x) { # x는 숫자형 벡터
  return( x < summary(x)[2] - 1.5 * IQR(x) | x > summary(x)[5] + 1.5 * IQR(x) )
}#outlier 이상만 줌.

# is_outlier()를 사용하여 outlier 데이터에만 is_out 필드에 타자이름 넣기
HR_outlier <- TB_2017 %>%
 group_by(team) %>%
 mutate(is_out = ifelse(is_outlier(HR), batter_name, NA))#is out에 이름이 있으면 홈런을 많이 친 사람.

print("HR_outlier에 HR이 outlier인 데이터에만 is_out 필드에 타자이름을 저장한 후")
HR_outlier %>% show()
```

####11. [확인] HR_outlier에서 HR(홈런)이 outlier에 해당하는 선수들의 정보를 출력하고geom_boxplot()으로 나타내는 방법을 확인하세요.
(Hint) ggthemes 라이브러리

```r
#11 [확인] HR_outlier에서 HR(홈런)이 outlier에 해당하는 선수들의 정보를  출력하고
# geom_boxplot()으로 나타내는 방법을 확인하세요. (Hint) ggthemes 라이브러리

ggplot(data = HR_outlier, mapping = aes(x = team, y = HR)) +
geom_boxplot(mapping = aes(fill = team), outlier.color = "red", outlier.shape = "★", outlier.size = 7) +
geom_text(mapping = aes(label = is_out), na.rm = T, vjust = -1) +
theme(axis.title.x = element_text(family = fonts()[1])) + # 한글 설정.
theme_wsj()
```

####13. TB 데이터에서 숫자인 데이터가 NA인 개수를 확인한 후 숫자인 데이터의 NA를 모두 0으로바꿔 TB_NAto0에 저장하고확인하세요.

```r
#13 TB 데이터에서 숫자인 데이터가 NA인 개수를 확인한 후 숫자인 데이터의 NA를 모두 0으로바꿔 TB_NAto0에 저장하고 확인하세요.
#  (Hint) select_if(is.numeric), replace(), is.na(.)

TB %>% select_if(is.numeric) %>% is.na() %>% table()

TB_NAto0 <- TB %>% select_if(is.numeric) %>% replace(is.na(.),0)

TB_NAto0 %>% select_if(is.numeric) %>% is.na() %>% table()
```

```r
#15 TB 데이터와 TB_NAto0 데이터의 5열부터 12열까지의 상관계수를 소수 둘째 자리까지 각각
#   TB_cor과 TB_NAto0_cor에 저장한 후 가장 상관관계가 높은 필드를 제시하세요.
TB_cor <- TB[,5:12] %>% cor() %>% round(2)
TB_cor

TB_NAto0_cor <- TB_NAto0[,5:12] %>% cor() %>% round(2)
TB_NAto0_cor
```
